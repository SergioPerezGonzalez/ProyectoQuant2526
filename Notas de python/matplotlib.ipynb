Debajo hay un ejemplo creado con IA para visualizar lo que hace mosaic, es una forma de gráficas con una composición muy personalizada, sintaxis en el punto 2 y 3

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generamos algunos datos de prueba
x = np.linspace(0, 10, 100)
y_seno = np.sin(x)
x_barras = ['A', 'B', 'C', 'D']
y_barras = [10, 20, 15, 25]
x_puntos = np.random.rand(50)
y_puntos = np.random.rand(50)

# 2. Definimos el mapa (el "mosaico")
mapa = [
    ['left', 'right_top'],
    ['left', 'right_bottom']
]

# 3. Creamos la figura y los ejes usando el mapa
fig, axs = plt.subplot_mosaic(mapa, figsize=(10, 6))

# 4. Dibujamos en la gráfica de la IZQUIERDA ('left')
axs['left'].plot(x, y_seno, color='blue', linewidth=2)
axs['left'].set_title('Gráfica Principal (Seno)')
axs['left'].set_xlabel('Eje X')
axs['left'].set_ylabel('Amplitud')

# 5. Dibujamos en la gráfica ARRIBA A LA DERECHA ('right_top')
axs['right_top'].scatter(x_puntos, y_puntos, color='red', alpha=0.6)
axs['right_top'].set_title('Gráfico de Dispersión')

# 6. Dibujamos en la gráfica ABAJO A LA DERECHA ('right_bottom')
axs['right_bottom'].bar(x_barras, y_barras, color='green')
axs['right_bottom'].set_title('Gráfico de Barras')

# 7. Ajustamos el espaciado para que no se superpongan los títulos con los ejes
plt.tight_layout()

# 8. Mostramos el resultado final
plt.show()
```

Anatomía de una tabla

Los datos y textos

```python
conjunto = [0, 1, 2, 3]
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
#OJO: label es simplemente un nombre para la tabla
#OJO: con varios subplots, lo que hay que hacer es indicar el axis (tabla) donde estas trabajando
ax[0][0].plot(conjunto, conjunto, color='red', linewidth=5, label='Tabla 1') #plot = lineal
#OJO: linewidth es para el grosor de las lineas, en otras tablas, es para los contornos
#marker='s' es para cuadrados (squares), 'o' para círculos
ax[1][0].scatter(conjunto, conjunto, color='purple', marker='s', alpha=1, label='tabla 2') #scatter es de puntos dispersos
#OJO: alpha es la opacidad
ax[0][1].bar(conjunto, conjunto, color='yellow',  edgecolor='black', linewidth=10, alpha=1, label='tabla 3') #categorias, valores
#OJO: mismos elementos personalizables que las demás tablas
ax[1][1].barh(conjunto, conjunto, color='yellow',  edgecolor='black', linewidth=10, alpha=1, label='tabla 3') #categorias, valores
#OJO: como .bar pero en horizontal
```

```python
conjunto = [0, 1, 2, 3]
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0][0].hist(conjunto, color='red', edgecolor='black', linewidth=5, label='Tabla 1', bins=3) #OJO: solo un argumento
#bins son el número de cajas en la que se agrupa la información

ax[1][0].scatter(conjunto, conjunto, color='purple', marker='s', alpha=1, label='tabla 2') 

ax[0][1].bar(conjunto, conjunto, color='yellow',  edgecolor='black', linewidth=10, alpha=1, label='tabla 3') 

ax[1][1].barh(conjunto, conjunto, color='yellow',  edgecolor='black', linewidth=10, alpha=1, label='tabla 3') 
```
