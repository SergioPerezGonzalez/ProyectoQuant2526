Bucles for y funciones for:

```python
#Los for en python son todos for each
lista = [0, 1, 2]
for x in lista:
    print(x)
```

```python
#Funciones relacionadas con for:
n = 10
intervalo = 1
#OJO: SOLO SIRVEN PARA LAS FUNCIONES
crearListaNumeros = range(0, n, intervalo) #Crea una lista de numero desde 0 a n con intervalos intervalo, por defecto (0, n, 1)
crearEnumeración = enumerate(lista) #Enumera los elementos de una lista
len(lista) #length lista
zip(lista, lista) #producto cartesiano
sorted(lista)
reversed(lista) #orden invertido
```

Matemáticas básicas

```python
a = [0, 1, 2, 3]
max(a)
min(a)
sum(a) #sumatorio
b= a[1]
round(b) #redondear
abs(b) #valor absoluto
pow(b, b) #numero, elevado a
divmod(b, b) #dividendo, divisor. Devuelve cociente, resto

```

Funciones listas
