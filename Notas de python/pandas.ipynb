```python
import pandas as pd 

import kagglehub

# Download latest version
path = kagglehub.dataset_download("jasonqhiu/sp500")

print("Path to dataset files:", path)

df = pd.read_csv(f"{path}/GSPC.csv") #más rápido
```

Limpieza de datos

```python
tablaBooleanaNan = df.isna() #Tabla si hay NaN o no
tablaBooleanaNan.sum() #sumatorio para dataframes
df.dropna() #Adios datos vacíos
df.fillna("Hola pana") #los NaN se le añade el valor puesto 
```

```python
df.duplicated() #estan las filas duplicadas?
df.drop_duplicates() #adios filas duplicadas
```

```python
df['Date'].str.lower() #pasar a minúscula la columna entera
df['Date'].str.lower() #pasar a mayuscula la columna entera
df['Date'].str.strip() #quitar los espacios al principio y final
df['Date'].str.replace("/", ".") #cambiar los / por .
```

```python
df.rename(columns={'Date': "Fecha_bro"}, inplace=True) #OJO: nombre viejo, nombre nuevo 
#OJO: inplace: reemplazo el nombre de la original o no?
df.replace (0, 1) #elementos se ponen como lo diga el tipo de dato
df["Fecha_bro"] = df["Fecha_bro"].astype(str)
```
